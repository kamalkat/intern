```python
import pandas as pd

```

```python
df=pd.read_csv('/content/Unemployment in India.csv')
df.head()
```

```python
df=df.rename(columns={df.columns[3]:'UR',df.columns[4]:'E',df.columns[5]:'P'})
df
```

```python
df.isnull().sum().reset_index()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df=df.dropna(axis=0)
df.isnull().sum()
df.dropna()
```

```python
df.fillna(0,inplace=True)
df.fillna(df.mean(),inplace=True)
df
```

```python
import seaborn as sns

heat_maps=df[['UR','E','P']]
heat_maps=heat_maps.corr()
sns.heatmap(heat_maps,annot=True,cmap='winter')
print(df)
```

```python
sns.histplot(df,x='UR',hue='Region')
```

```python
sns.boxplot(df,x='UR')
```

```python
sns.violinplot(df,x='UR')
```
